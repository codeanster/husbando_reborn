```python
#compare cropped images to uncropped images, look at tags
import pandas as pd
import matplotlib.pyplot as plt
import glob as glob
import random
import matplotlib.image as mpimg
```

```python
file_path =  '/content/drive/MyDrive/husbando_reborn/data/cropped'
filelist = glob.glob(f'{file_path}/*')
```

```python
#let's start by picking a random id
ids = []
for i in range(len(filelist)):
  id = filelist[i].split('/')[-1].split('.')[0]
  ids.append(id)
```

```python
def plot_random_id(ids):
  try:
    random_id = random.choice(ids)
    #let's plot these two images side by side
    cropped_file_path = '/content/drive/MyDrive/husbando_reborn/data/cropped'
    uncropped_file_path = '/content/drive/MyDrive/husbando_reborn/data/uncropped'

    img1 = mpimg.imread(f'/content/drive/MyDrive/husbando_reborn/data/cropped/{random_id}.png')[:,:,::-1] #this correctly fixes how the image loooks.. will need to edit cropping script to include this change
    img2 = mpimg.imread(f'/content/drive/MyDrive/husbando_reborn/data/uncropped/{random_id}.png',0)


    #read in pandas dataframe
    df = pd.read_csv('/content/drive/MyDrive/husbando_reborn/meta.csv')


    # Creating 2 subplots
    fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(10,10))

    #tags
    tags = df[df['id'] == int(random_id)]['tags'].iloc[0]
    
    # Accessing each axes object to plot the data through returned array
    ax[0].imshow(img1)
    ax[0].set_title('After Cropping')

    ax[1].imshow(img2)
    ax[1].set_title('Before Cropping')

    print(f'{tags}')
  except:
    pass
```

```python
plot_random_id(ids)
```
